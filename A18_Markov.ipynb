```python
import numpy as np
import matplotlib.pyplot as plt

# Impostazioni della passeggiata casuale
n_steps = 1000  # Numero di passi
p = 0.5         # Probabilità di andare a destra
q = 1 - p       # Probabilità di andare a sinistra

# Genera i passi (1 per destra, -1 per sinistra)
steps = np.random.choice([-1, 1], size=n_steps, p=[p, q])

# Calcola le posizioni
positions = np.cumsum(steps)

# Grafico del risultato
plt.figure(figsize=(10, 6))
plt.plot(positions)
plt.title("Passeggiata Casuale Semplice")
plt.xlabel("Passi")
plt.ylabel("Posizione")
plt.show()
```
