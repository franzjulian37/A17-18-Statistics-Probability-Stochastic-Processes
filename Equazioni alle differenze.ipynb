**Modelli Dinamici Discreti**

Definizione: I modelli dinamici discreti descrivono l'evoluzione di un sistema nel tempo attraverso iterazioni discrete, spesso utilizzando equazioni ricorsive (equazioni alle differenze finite).


Iterazioni discrete: Il tempo avanza di unità discrete (es., anni, generazioni).
Stabilità e comportamento asintotico: Studio dei punti fissi e della loro stabilità.
Applicazioni: Biologia (dinamica delle popolazioni), economia (modelli di crescita), ecologia, fisica, ecc.

**Equazioni alle Differenze **

Definizione: Le equazioni alle differenze finite sono equazioni ricorsive che definiscono una sequenza numerica basata sui valori precedenti.

**Modello Base delle Popolazioni di Malthus**

Contesto Storico: Proposto da Thomas Robert Malthus nel 1798, il modello di Malthus descrive la crescita di una popolazione in assenza di limiti ambientali.

```python
import matplotlib.pyplot as plt

def malthus_model(P0, r, periods):
    population = [P0]
    for t in range(1, periods + 1):
        Pt = (1 + r) * population[-1]
        population.append(Pt)
    return population

# Parametri
P0 = 100       # Popolazione iniziale
r = 0.05       # Tasso di crescita (5%)
periods = 50   # Numero di periodi

# Simulazione
population = malthus_model(P0, r, periods)

# Visualizzazione
plt.plot(range(periods + 1), population, marker='o')
plt.title("Modello di Malthus per la Crescita della Popolazione")
plt.xlabel("Tempo (periodi)")
plt.ylabel("Popolazione")
plt.grid(True)
plt.show()
```

I modelli dinamici discreti e le equazioni alle differenze finite sono strumenti potenti per descrivere e analizzare l'evoluzione di sistemi nel tempo. Il modello di Malthus rappresenta un caso base di crescita esponenziale, utile per comprendere concetti fondamentali come tasso di crescita e comportamento asintotico. Attraverso l'analisi e la simulazione, è possibile esplorare diverse dinamiche popolazionali e estendere il modello per includere fattori reali come risorse limitate.

Il modello di Malthus può essere esteso per includere limiti alla crescita, portando al modello logistico dove K è la capacità portante dell'ambiente. Questo introduce un elemento di stabilità, permettendo alla popolazione di stabilizzarsi a K.

```python
import matplotlib.pyplot as plt

def logistic_model(P0, r, K, periods):
    population = [P0]
    for t in range(1, periods + 1):
        Pt = population[-1] + r * population[-1] * (1 - population[-1] / K)
        population.append(Pt)
    return population

# Parametri
P0 = 10        # Popolazione iniziale
r = 0.2        # Tasso di crescita (20%)
K = 100        # Capacità portante
periods = 50   # Numero di periodi

# Simulazione
population = logistic_model(P0, r, K, periods)

# Visualizzazione
plt.plot(range(periods + 1), population, marker='o')
plt.axhline(y=K, color='r', linestyle='--', label=f"Capacità portante K={K}")
plt.title("Modello Logistico per la Crescita della Popolazione")
plt.xlabel("Tempo (periodi)")
plt.ylabel("Popolazione")
plt.legend()
plt.grid(True)
plt.show()
```
